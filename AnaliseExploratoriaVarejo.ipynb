# Pergunta de Negócio 8 (Desafio Nível Master):

**Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. **

Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
df['Valor_Desconto'] = df['Valor_Venda'] - (df['Valor_Venda'] * df['Desconto'])
```

```python
df.head()
```

```python
# Filtrando as vendas antes do desconto de 15%
vendas_antes_desconto = df.loc[df['Desconto'] == 0.15, 'Valor_Venda']
```

```python
# Filtrando as vendas depois do desconto de 15%
vendas_depois_desconto = df.loc[df['Desconto'] == 0.15, 'Valor_Desconto']
```

```python
# Calcula a média das vendas antes do desconto de 15%
media_vendas_antes_desconto = vendas_antes_desconto.mean()
```

```python
# Calcula a média das vendas depois do desconto de 15%
media_vendas_depois_desconto = vendas_depois_desconto.mean()
```

```python
print("Média das vendas antes do desconto de 15%:", round(media_vendas_antes_desconto, 2))
```

```python
print("Média das vendas depois do desconto de 15%:", round(media_vendas_depois_desconto, 2))
```

# Pergunta de Negócio 9 (Desafio Nível Master Ninja):

**Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?**

Demonstre o resultado através de gráfico de linha.

```python
df['Data_Pedido'] = pd.to_datetime(df['Data_Pedido'], format='%d/%m/%Y')
df['Mes'] = df['Data_Pedido'].dt.month
df['Ano'] = df['Data_Pedido'].dt.year
```

```python
# Agrupamos por ano, mês e segmento e calculamos estatísticas de agregação
agrup = df.groupby(['Ano', 'Mes', 'Segmento'])['Valor_Venda'].agg([np.sum, np.mean, np.median])
```

```python
agrup
```

```python
anos = agrup.index.get_level_values(0)
meses = agrup.index.get_level_values(1)
segmentos = agrup.index.get_level_values(2)
```

```python
# Plot
mpl.figure(figsize = (12, 6))
sb.set()
fig1 = sb.relplot(kind = 'line',
                   data = agrup,
                   y = 'mean',
                   x = meses,
                   hue = segmentos,
                   col = anos,
                   col_wrap = 4)
mpl.show()
```

# Pergunta de Negócio 10 (Desafio Nível Master Ninja das Galáxias):

**Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias?**

Demonstre tudo através de um único gráfico.

```python
agrup = df.groupby(['Categoria',
                             'SubCategoria']).sum(numeric_only = True).sort_values('Valor_Venda',
                                                                                   ascending = False).head(12)
```

```python
agrup  = agrup[['Valor_Venda']].astype(int).sort_values(by = 'Categoria').reset_index()
```

```python
agrup
```

```python
agrup_cat = agrup.groupby('Categoria').sum(numeric_only = True).reset_index()
```

```python
agrup_cat
```

```python
# Listas de cores para categorias
cores_categorias = ['#5d00de',
                    '#0ee84f',
                    '#e80e27']
```

```python
# Listas de cores para subcategorias
cores_subcategorias = ['#aa8cd4',
                       '#aa8cd5',
                       '#aa8cd6',
                       '#aa8cd7',
                       '#26c957',
                       '#26c958',
                       '#26c959',
                       '#26c960',
                       '#e65e65',
                       '#e65e66',
                       '#e65e67',
                       '#e65e68']
```

```python
# Define autopct_format function
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ($ {v:d})'.format(p=pct,v=val)
    return my_format

# Plot

# Tamanho da figura
fig, ax = mpl.subplots(figsize = (18,12))

# Gráfico das categorias
p1 = ax.pie(agrup_cat['Valor_Venda'],
            radius = 1,
            labels = agrup_cat['Categoria'],
            wedgeprops = dict(edgecolor = 'white'),
            colors = cores_categorias)

# Gráfico das subcategorias
p2 = ax.pie(agrup['Valor_Venda'],
            radius = 0.9,
            labels = agrup['SubCategoria'],
            autopct = autopct_format(agrup['Valor_Venda']), # Calling the defined function
            colors = cores_subcategorias,
            labeldistance = 0.7,
            wedgeprops = dict(edgecolor = 'white'),
            pctdistance = 0.53,
            rotatelabels = True)

# Limpa o centro do círculo
centre_circle = mpl.Circle((0, 0), 0.6, fc = 'white')

# Labels e anotações
fig = mpl.gcf()
fig.gca().add_artist(centre_circle)
mpl.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(agrup['Valor_Venda']))), xy = (-0.2, 0))
mpl.title('Total de Vendas Por Categoria e Top 12 SubCategorias')
mpl.show()
```
